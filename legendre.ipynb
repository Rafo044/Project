```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.special import legendre  # Yüksəklik məlumatı ( dəniz səviyyəsinə görə fərqlilikləri) 
#  100 nöqtə var, -50 m ilə +50 m arasında dəyişir
yukseklikler = np.linspace(-50, 50, 100)  # Legendre polinomunun dərəcəsi 
derece = 3  # Legendre polinomlarını hesablanması
polinomlar = legendre(derece)
# Gravitasiya sabiti 
gravitasiya_sabiti = 9.81   
# Yerin cazibə qüvvəsini hesablanması 
gravity = polinomlar(yukseklikler) * gravitasiya_sabiti 
print(gravity)
# Vizualizasiya
plt.plot(yukseklikler, gravity,color = "red") 
plt.xlabel('Yüksəklik (m)') 
plt.ylabel('Təcil (m/s^2)') 
plt.title('Yüksəkliklərə görə gravitasiya təcili') 
plt.grid(True) 
plt.show()
```
