```python
import numpy as np
from scipy.signal import savgol_filter  
# Data 
x = np.linspace(0, 2*np.pi, 100) 
y = np.sin(x) + np.random.normal(0, 0.1, 100)  # Sinus dalğası  
# Svitzky-Golay filtri
window_size = 5  # Pəncərə ölçüsü
poly_order = 2   # Polinom dərəcəsi 
smoothed_data = savgol_filter(y, window_size, poly_order)  
# Matplotliblə vizualizasiya
import matplotlib.pyplot as plt  
plt.figure(figsize=(8, 6)) 
plt.plot(x, y, label="Səs-küylü(noisy data)")
plt.plot(x, smoothed_data, label="Düzləştirilmiş data", 
		 color='red', linewidth=2) 
plt.legend() 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title("Svitzky-Golay Filtri ilə Düzləştirmə") 
plt.grid(True) 
plt.show()
```
