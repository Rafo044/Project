```python
import numpy as np 
import matplotlib.pyplot as plt
#Polar grafik üçün datalar 
theta = np.linspace(0, 2*np.pi, 6, endpoint=False)
radii = [2, 3, 1, 4, 2, 3] 
# Polar grafik vizualizasiyası 
plt.figure(figsize=(6, 6))
ax = plt.subplot(2, 1, 1, polar=True)
ax.plot(theta, radii, marker='o') 
ax.set_title('Polar Grafiki') 
plt.show()
```

```python
from wordcloud import WordCloud 
import matplotlib.pyplot as plt  
# Metin verisi 
text = "Hello  world! Python wordcloud,python."  
# Kelime bulutu oluşturma 
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)  
# Kelime bulutu çizimi 
plt.figure(figsize=(10, 5)) 
plt.imshow(wordcloud, interpolation='bilinear') 
plt.axis('off') 
plt.title('Example box') 
plt.show()
```
