```python
import numpy as np

# Polinom əmsalları
coefficient = [1, -2, 1]  # Polinom: x^2 - 2x + 1

# Hər hansı bir x dərəri 
x = 2

# Polinomu değerlendirme
y = np.polyval(coefficient, x)
print("Result:", y)  # Output: 1
```

```python
import numpy as np

# Datalar
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2, 1, 3, 5, 4])

# Polinom tətbiqi
coefficient = np.polyfit(x_data, y_data, 2)  # 2. dərəcədən polinom

print("Polinom əmsalları:", coefficient)  # Output: [ 0.2 -0.9  1.4]
```

```python
# Matrislər
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Matris hasili
C = np.dot(A, B)
print("Matris hasili:", C)

# Matrisin tərsi
A_ters = np.linalg.inv(A)
print("Matris Tərsi:", A_ters)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Datalar
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2, 1, 3, 5, 4])

# Polinom uyumlama
coefficient = np.polyfit(x_data, y_data, 2)  # 2. dərəcədən polinom

# Polinoma grafik interval təyin edib dəyərləndirək
x_interval = np.linspace(0, 6, 100)  # Grafik interval
y_interval_deyeri = np.polyval(coefficient, x_interval)

# Məlumatları qrafikdə vizuallaşdırma
plt.scatter(x_data, y_data, color='red', label='Veriler')
plt.plot(x_interval, y_interval_deyeri, color='blue',label='Polinom')
plt.xlabel('X Dəyərləri')
plt.ylabel('Y Dəyərləri')
plt.title('Bütün datalar')
```
